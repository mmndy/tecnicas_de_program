```python
#1. Escreva um programa que solicita que o usuário digite 5 números inteiros e os coloca em uma lista. Quando terminar, imprima a lista:

numeros = []

for i in range(5):
    numero = int(input(f"Digite o {i+1}º número inteiro: "))
    numeros.append(numero)

print("A lista de números digitados é:", numeros)
```

```python
#2. Escreva um programa que aplica a função módulo a todos os elementos e uma lista.

numeros = [int(x) for x in input("Digite números separados por espaço: ").split()]

modulos = [abs(x) for x in numeros]

print("Lista com módulo aplicado:", modulos)
```

```python
#3.Escreva um programa que apaga todos os elementos negativos de uma lista.

numeros = [int(x) for x in input("Digite números separados por espaço: ").split()]

positivos = [x for x in numeros if x >= 0]

print("Lista sem números negativos:", positivos)
```

```python
#4. Faça um programa que solicita uma quantidade indefinida de notas de uma prova (deve parar quando um número negativo for digitado). Armazena em uma lista.Após isso, o programa deve mostrar a média das notas.

notas = []

while True:
    nota = float(input("Digite uma nota (ou um número negativo para sair): "))

    if nota < 0:
        break

    notas.append(nota)

if len(notas) > 0:
    media = sum(notas) / len(notas)
    print("Média das notas:", media)
else:
    print("Nenhuma nota foi digitada.")
```

```python
#5. Faça um programa que solicita uma quantidade indefinida de números positivos (deve parar quando um número negativo for
#digitado). Armazene-os em duas listas: uma para os números pares; outra para os números ímpares. A seguir, mostre a porcentagem de pares e ímpares digitados.

pares = []
impares = []

while True:
    numero = int(input("Digite um número positivo (ou um número negativo para sair): "))

    if numero < 0:
        break

    if numero % 2 == 0:
        pares.append(numero)
    else:
        impares.append(numero)

total = len(pares) + len(impares)

if total > 0:
    perc_pares = (len(pares) / total) * 100
    perc_impares = (len(impares) / total) * 100

    print(f"Porcentagem de números pares: {perc_pares:.2f}%")
    print(f"Porcentagem de números ímpares: {perc_impares:.2f}%")
else:
    print("Nenhum número foi digitado.")
```

```python
#6. Faça um programa que solicita uma quantidade indefinida de números positivos (deve parar quando um número negativo for
#digitado). Armazene todos os números digitados em uma lista, sem repetição.

numeros = []

while True:
    numero = int(input("Digite um número positivo (ou um número negativo para sair): "))


    if numero < 0:
        break


    if numero not in numeros:
        numeros.append(numero)

print("Números digitados (sem repetição):", numeros)
```

```python
#7. Escreva um programa que solicita que o usuário digite a altura e o peso de 5 pessoas. Verifique se pelo menos duas das pessoas tem a mesma altura e mesmo peso.

pessoas = []

for i in range(5):
    print(f"\nPessoa {i+1}:")
    altura = float(input("Digite a altura (em metros): "))
    peso = float(input("Digite o peso (em kg): "))

    pessoas.append((altura, peso))

duplicado = False
for i in range(len(pessoas)):
    for j in range(i + 1, len(pessoas)):
        if pessoas[i] == pessoas[j]:
            duplicado = True
            break

if duplicado:
    print("\nPelo menos duas pessoas têm a mesma altura e o mesmo peso.")
else:
    print("\nNenhuma dupla de pessoas tem a mesma altura e peso.")
```

```python
#8. Faça um programa que solicita que o usuário digite o nome e a idade de diversas pessoas (o programa deve parar quando um nome vazio for informado).
#Retorne o nome da pessoa mais velha.

pessoas = []

while True:
    nome = input("Digite o nome da pessoa (ou pressione Enter para encerrar): ")


    if nome == "":
        break

    idade = int(input("Digite a idade: "))
    pessoas.append((nome, idade))


if len(pessoas) > 0:
    mais_velha = max(pessoas, key=lambda p: p[1])
    print(f"\nA pessoa mais velha é {mais_velha[0]}, com {mais_velha[1]} anos.")
else:
    print("\nNenhuma pessoa foi cadastrada.")
```
