```python
#7. Solicite ao usuário três notas e seus respectivos pesos, depois calcule e exiba a média ponderada.

n1 = float(input("Digite a primeira nota:  "))
peso1 = float(input("Digite o peso da primeira nota:  "))

n2 = float(input("Digite a segunda nota:  "))
peso2 = float(input("Digite o peso da segunda nota:   "))

n3 = float(input("Digite a terceira nota:  "))
peso3 = float(input("Digite o peso da terceira nota:  "))


media_ponderada = (n1*peso1 + n2*peso2 + n3*peso3)/ (peso1 + peso2 + peso3)

print(f"A média ponderada é:{media_ponderada: .2f}")
```

```python
#8. Peça ao usuário um número e exiba a soma de seus dígitos.

numero = input("Digite um número: ")
soma = 0

for digito in numero:
  soma += int(digito)

print(f"A soma dos números é: {soma}")
```

```python
#9. Peça ao usuário uma palavra e exiba-a. invertida sem usar laços.

palavra = input("Digite uma palavra:")
palavra_invertida = palavra[:: -1]
print(f"A palavra invertida é: {palavra_invertida}")
```
