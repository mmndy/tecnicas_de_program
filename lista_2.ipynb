```python
#10. O sistema de avaliação de determinada disciplina é composto por três provas. A primeira
#prova tem peso 2, a segunda tem peso 3 e a terceira tem peso 5. Considerando que a
#média para aprovação é 6.0, faça um algoritmo para calcular a média final de um aluno
#desta disciplina e dizer se o aluno foi aprovado ou não

p1 = 2
p2 = 3
p3 = 5

nota1 = float(input("Digite sua nota na P1: "))
nota2 = float(input("Digite sua nota na P2: "))
nota3 = float(input("Digite sua nota na P3: "))

media = (nota1*p1 + nota2*p2 + nota3*p3) / (p1 + p2 + p3)

print("Média:", media)

if media >= 6:
    print("Aluno aprovado!")

else:
    print("Reprovado!")







```

```python
#14. Escreva um programa em linguagem C que solicite ao usuário a média para aprovação
#em um curso e em seguida solicite ao usuário o nome, sexo e as 03 notas do aluno e ao
#final imprima a frase: "O aluno XXXXX foi aprovado com media YY" considerando o gênero
#do(a) aluno(a) e se foi aprovado(a) ou reprovado(a)

#include <stdio.h>

int main() {
    float media_aprovacao;
    char nome[50];
    char sexo;
    float nota1, nota2, nota3;
    float media_aluno;

    printf("Digite a media para aprovacao: ");
    scanf("%f", &media_aprovacao);

    printf("Digite o nome do aluno: ");
    scanf(" %49[^\n]", nome);  // lê até a quebra de linha, até 49 caracteres

    printf("Digite o sexo do aluno (M/F): ");
    scanf(" %c", &sexo);

    printf("Digite as 3 notas do aluno:\n");
    printf("Nota 1: ");
    scanf("%f", &nota1);
    printf("Nota 2: ");
    scanf("%f", &nota2);
    printf("Nota 3: ");
    scanf("%f", &nota3);

    media_aluno = (nota1 + nota2 + nota3) / 3.0;


    if (media_aluno >= media_aprovacao) {
        if (sexo == 'M' || sexo == 'm') {
            printf("O aluno %s foi aprovado com media %.2f\n", nome, media_aluno);
        } else if (sexo == 'F' || sexo == 'f') {
            printf("A aluna %s foi aprovada com media %.2f\n", nome, media_aluno);
        } else {
            printf("Aluno(a) %s foi aprovado(a) com media %.2f\n", nome, media_aluno);
        }
    } else {
        if (sexo == 'M' || sexo == 'm') {
            printf("O aluno %s foi reprovado com media %.2f\n", nome, media_aluno);
        } else if (sexo == 'F' || sexo == 'f') {
            printf("A aluna %s foi reprovada com media %.2f\n", nome, media_aluno);
        } else {
            printf("Aluno(a) %s foi reprovado(a) com media %.2f\n", nome, media_aluno);
        }
    }

    return 0;
}
```
